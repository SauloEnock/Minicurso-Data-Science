```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline

import warnings
warnings.filterwarnings('ignore')

seed_number = 123
```

```python
# VERIFICAÇÃO DOS DADOS
```

```python
train = pd.read_csv('input/train.csv')
test = pd.read_csv('input/test.csv')
train.shape, test.shape
```

```python
train.head()
```

```python
train.set_index('Id', inplace=True)
test.set_index('Id', inplace=True)
```

```python
test.head()
```

```python
data = pd.concat([train,test],sort=True)
data.shape, data.columns
```

```python
data.describe()
```

```python
data.describe(include=['O'])
```

```python
# LIMPEZA DOS DADOS
```

```python
# Retirada das colunas pois possuiem muitas variações.
data.drop(['Date', 'Location', 'WindDir3pm', 'WindDir9am', 'WindGustDir'], axis=1, inplace=True)
data.shape
```

```python
data.head(100)
```

```python
data.isnull().sum()
```

```python
# Retirada dessas colunas pois apresentam 40% ou mais do seu total em valores Nulos
data.drop(['Cloud3pm', 'Cloud9am', 'Evaporation', 'Sunshine'], axis=1, inplace=True)
data.shape
```

```python
# As colunas a seguir foram mantidas pois possuiam uma porcentagem aceitável de valores nulos  

#Preencher RainToday (Categórico) com Moda
data['RainToday'].fillna(data['RainToday'].mode()[0], inplace=True)

#Preencher Humidity3pm (Contínuo) com Mediana
data['Humidity3pm'].fillna(data['Humidity3pm'].median(), inplace=True)

#Preencher Humidity9am (Contínuo) com Mediana
data['Humidity9am'].fillna(data['Humidity9am'].median(), inplace=True)

#Preencher MaxTemp (Contínuo) com Média
data['MaxTemp'].fillna(data['MaxTemp'].mean(), inplace=True)

#Preencher MinTemp (Contínuo) com Média
data['MinTemp'].fillna(data['MinTemp'].mean(), inplace=True)

#Preencher Pressure3pm (Contínuo) com Média
data['Pressure3pm'].fillna(data['Pressure3pm'].mean(), inplace=True)

#Preencher Pressure9am (Contínuo) com Média
data['Pressure9am'].fillna(data['Pressure9am'].mean(), inplace=True)

#Preencher Rainfall (Contínuo) com Média
data['Rainfall'].fillna(data['Rainfall'].mean(), inplace=True)

#Preencher Temp3pm (Contínuo) com Média
data['Temp3pm'].fillna(data['Temp3pm'].mean(), inplace=True)

#Preencher Temp9am (Contínuo) com Média
data['Temp9am'].fillna(data['Temp9am'].mean(), inplace=True)

#Preencher WindGustSpeed (Contínuo) com Média
data['WindGustSpeed'].fillna(data['WindGustSpeed'].median(), inplace=True)

#Preencher WindSpeed3pm (Contínuo) com Média
data['WindSpeed3pm'].fillna(data['WindSpeed3pm'].median(), inplace=True)

#Preencher WindSpeed9am (Contínuo) com Média
data['WindSpeed9am'].fillna(data['WindSpeed9am'].median(), inplace=True)

data.isnull().sum()
```
